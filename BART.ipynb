# BART Embedding

```python
def bart_embedding(df, text_col, batch_size=1):
    # 모델 이름 설정 및 로드
    model_name = 'facebook/bart-base'
    tokenizer = BartTokenizer.from_pretrained(model_name)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = BartModel.from_pretrained(model_name).to(device)

    embeddings = []
    text_list = df[text_col].tolist()      # df[text_col]을 모두 한 list로 합치기

    # 텍스트 리스트를 배치로 처리하여 임베딩 생성
    for i in tqdm(range(0, len(text_list), batch_size)):
        batch_reviews = text_list[i:i + batch_size]
        inputs = tokenizer(batch_reviews, return_tensors='pt', truncation=True)      # truncation : 데이터 자르기

        input_ids = inputs["input_ids"].to(device)                # input_ids : 텍스트 입력을 토큰으로 변환한 후, 각 토큰을 정수 ID로 매핑
        attention_mask = inputs["attention_mask"].to(device)      # attention_mask :  실제 토큰 1, 패딩 토큰 0으로 구분

        # PyTorch에서의 추론 수행
        with torch.no_grad():  # 추론 시에는 gradient 계산을 비활성화
            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            # BART 모델에서 임베딩 추출
            encoder_hidden_states = outputs.last_hidden_state

            # 각 입력의 마지막 hidden state 선택 (일반적으로 분류 작업에 사용됨)
            result = encoder_hidden_states[:, -1, :][0].cpu().numpy()
            embeddings.append(result)

    stacked_embeddings = np.vstack(embeddings)      # vstack : 수직으로 결합
    df['bart_embedding']= list(stacked_embeddings)

    return df
```
